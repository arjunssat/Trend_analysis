```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from pytrends.request import TrendReq
trends=TrendReq()
```

```python
trends.build_payload(kw_list=['S400'])
data=trends.interest_by_region()
data=data.sort_values(by='S400',ascending=False)
data=data.head(10)
print(data)
```

```python
ax=data.reset_index().plot(x='geoName',y='S400',figsize=(30,15),kind='barh')
plt.style.use('ggplot')
ax.tick_params(axis='both', which='major', labelsize=25)
plt.show()
```

```python
trends.build_payload(kw_list=['russia S400'])
related=trends.related_queries()
related.values()
```

```python
data=TrendReq(hl='en-US',tz=300)
data.build_payload(kw_list=['russia S400'])
data=data.interest_over_time()
fig,ax=plt.subplots(figsize=(20,10))
data['russia S400'].plot()
plt.style.use('dark_background')
plt.title('Total Google searches for S400',fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

The above analysis shows that the demand for Russian-S400 in google searches were at peak(100mn) during the period of mid-2020 and the search rate dropped significantly after 2020 and the demand for S400 raised again to 80mn by early 2022 and further dropped afterwards.Further the query trend analysis shows that there was a huge trend for India-Russia S400 deal,china purchase of S400 and Ukraine-Russia war impact of S400.Also among those Turkey,Jordan,Algeria & Cyprus searched for S400 most times than any other country.
